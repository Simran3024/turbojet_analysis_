```python
#End of life (EOL)  tell the last index 
EOL=[]
for sl_no in raw_data['ID']:
    EOL.append( ((raw_data[raw_data['ID'] == sl_no]["Cycle"]).values)[-1])    
EOL
```

```python
raw_data['EOL']=EOL
```

```python
raw_data
```

```python
raw_data['LR']=raw_data['Cycle'].div(raw_data['EOL'])  #present age divide by end age  (LR=donot do this work)
```

```python
raw_data
```

```python
raw_data=raw_data.drop(columns=['ID','EOL'])
```

```python
raw_data
```

```python
labels=[]
for i in range(0,len(raw_data)):
    if np.array(raw_data['LR'])[i]<=0.6:
        labels.append(0)
    elif np.array(raw_data['LR'])[i]<=0.8:
        labels.append(1)
    else:
        labels.append(2)
raw_data['labels']=labels             
```
